# Feature Extraction 

## SIFT 

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('dataset/imgs/ship.jpg')
```

```python
plt.subplot(121)
plt.imshow(img)
plt.title('Original')

gray  = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
sift = cv2.SIFT_create()

kp = sift.detect(gray, None)

img = cv2.drawKeypoints(gray, kp, img)
plt.subplot(122)
plt.imshow(img)
plt.show()
```
